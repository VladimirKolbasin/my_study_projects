```python
# импорт библиотек numpy и pandas
import numpy as np
import pandas as pd

# считываем все csv файлы в датасеты
progress_phases = pd.read_csv('Data\progress_phases.csv')
progresses = pd.read_csv('Data\progresses.csv')
courses = pd.read_csv('Data\courses.csv')
course_contents = pd.read_csv('Data\course_contents.csv')

# объединяем датасеты
main = progress_phases.merge(progresses, how='left', left_on='progress_id', right_on='id')

# удаляем лишний столбец
main.pop('id')

# изменяем названия столбцов
courses1 = courses[['id', 'title', 'field']]
courses1.columns = ['id', 'course_title', 'course_field']
main = main.merge(courses1, how='left', left_on='course_id', right_on='id')

# удаляем лишний столбец
main.pop('id')
main = main.merge(course_contents, how='left', left_on=['module_number', 'course_id', 'lesson_number'],\
                  right_on=['module_number', 'course_id', 'lesson_number'])

# выводим общую информацию о получившемся датасете
main.info()
```

```python
# первых пять строк датасета main
main.head()
```

```python
# импорт модуля datetime для работы с датами
from datetime import datetime

# общее количество курсов в датасете: 15
main['course_id'].drop_duplicates().count()

# количество модулей на каждом курсе
number_of_modules = main[['course_title', 'module_title']].drop_duplicates().reset_index()\
    .groupby('course_title')['module_title'].count().reset_index()\
    .sort_values(by='module_title', ascending=False)
print('Количество модулей на каждом курсе:\n')
print(number_of_modules, '\n')

# количество уроков в каждом модуле на каждом курсе,
number_of_lessons = main[['course_title', 'module_title', 'lesson_title']].drop_duplicates().reset_index()\
    .groupby(['course_title', 'module_title'])['lesson_title'].count().reset_index()
print('Количество уроков в каждом модуле на каждом курсе:\n')
print(number_of_lessons.head(15), '\n')

# медианное количество уроков в модуле на каждом курсе
number_of_lessons_median = number_of_lessons.groupby('course_title')['lesson_title'].median().reset_index()\
    .sort_values(by='lesson_title', ascending=False)
print('Медианное количество уроков в модуле на каждом курсе:\n')
print(number_of_lessons_median, '\n')

# количество учеников на каждом курсе
number_of_students_by_course = main[['course_title', 'student_id']].drop_duplicates().reset_index()\
    .groupby('course_title')['student_id'].count().reset_index()\
    .sort_values(by='student_id', ascending=False)
print('Количество учеников на каждом курсе:\n')
print(number_of_students_by_course, '\n')
```

```python
# Возраст студентов считаем на дату начало учебы!!!

# минимальный, максимальный, средний, медианный возраст студентов
students = pd.read_csv('Data\students.csv', parse_dates=['birthday'], dayfirst=False)
students.columns = ['id_', 'student_id', 'city', 'birthday']

    # Для корректности данных - будем брать студентов, которые зачислены и начали проходить 1-й модуль, 1-й урок
    # на конктретный курс и удалим данные с пропущенными значениями
    # присвоим столбцам start_date, finish_date тип datetime
main['start_date'] = pd.to_datetime(main['start_date'])
main['finish_date'] = pd.to_datetime(main['finish_date'])
main_stud_course = main[['student_id', 'course_title', 'module_number', 'lesson_number', 'start_date']]
main_stud_course = main_stud_course[(main_stud_course['module_number'] == 1)\
                                    & (main_stud_course['lesson_number'] == 1)]

    # дополняем данные в датсет students из main, удаляем Nan
students_pure = students.merge(main_stud_course[['student_id', 'course_title','start_date']],\
                               how='left', left_on='student_id', right_on='student_id').dropna()

    #также очистим датасет от ошибок в датах и переведем в тип datetime стобец 'birthday':
students_pure['birthday'] = pd.to_datetime(students_pure['birthday'], format='%Y/%m/%d', errors='coerce')
students_pure = students_pure.dropna()

    # считаем количество лет, также возврат отсортируем от 10 до 100 лет
students_pure['age'] = students_pure['start_date'].dt.year - students_pure['birthday'].dt.year
students_pure = students_pure[(students_pure['age'] >= 10) & (students_pure['age'] <= 100)]

print('Минимальный возраст студентов:', students_pure['age'].min())
print('Максимальный возраст студентов:', students_pure['age'].max())
print('Средний возраст студентов:', round(students_pure['age'].mean()))
print('Медианный возраст студентов:', students_pure['age'].median())
students_pure['age'].describe()

# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

age_min_course = students_pure.groupby('course_title')['age'].min().reset_index().sort_values(by='age')
print('\nМинимальный возраст студентов в разрезе курсов\n')
print(age_min_course)

age_max_course = students_pure.groupby('course_title')['age'].max().reset_index().sort_values(by='age', ascending=False)
print('\nМаксимальный возраст студентов в разрезе курсов\n')
print(age_max_course)

age_mean_course = students_pure.groupby('course_title')['age'].mean().round().reset_index().sort_values(by='age')
print('\nСредний возраст студентов в разрезе курсов\n')
print(age_mean_course)

age_median_course = students_pure.groupby('course_title')['age'].median().reset_index().sort_values(by='age')
print('\nМедианный возраст студентов в разрезе курсов\n')
print(age_median_course)
```

```python
#Постройте bar-chart, отражающий количество студентов на каждом курсе. Ticks нужно развернуть так, чтобы они были читаемы

# Импорт библиотеки для построения графиков и диаграмм
import matplotlib.pyplot as plt

# Определение основных объектов диаграммы
fig, ax = plt.subplots(figsize=(12,6))

# Вывод графика на страницу
ax.bar(number_of_students_by_course['course_title'], number_of_students_by_course['student_id'])

# Подписание заголовка и осей координат
plt.title("Количество студентов на каждом курсе")
plt.ylabel('Количество студентов')
plt.xlabel('Курс')

# Определение угла наклона меток оси x
ax.tick_params(axis = 'x', rotation = 90)

# Отображение графика на странице
plt.show()
```

```python
#Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов
#на каждом курсе. График должен иметь заголовок. Значения должны быть отсортированы. Цвет столбцов должен содержать
#информацию о сфере, к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. 
#На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐

# импортируем seaborn для построения графиков 
import seaborn as sns

# выберем сколько студентов учится на каждом курсе (с учетом харакетиртиски сфера)
num_of_stud_by_course_field = main[['course_title','course_field', 'student_id']].drop_duplicates().reset_index()\
    .groupby(['course_title','course_field'])['student_id'].count().reset_index()\
    .sort_values(by='student_id')

# Определение основных объектов диаграммы
fig, ax = plt.subplots(figsize=(6,6), sharey=True)

# Вывод графика на страницу
sns.barplot(data=num_of_stud_by_course_field, x='student_id', y='course_title', ax=ax, ci=None, hue='course_field',\
            alpha=0.1, dodge=False)

# Подписание заголовка и осей координат
ax.set_title('Количество студентов на каждом курсе')
ax.set_ylabel('')
ax.set_xlabel('Количество студентов')

# строим медиану
ax.axvline(x=num_of_stud_by_course_field['student_id'].median(), color='red', linewidth=1, linestyle='--')

# выводим легенду
ax.legend()

# Отображение графика на странице
plt.show()
```

```python
#Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов,
#приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле),
#к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

# готовим датсет для подсчета конверсии
conversion_students = main[main['is_homework']==True]\
    [['course_title', 'course_field', 'module_number', 'module_title', 'student_id', 'start_date', 'finish_date']]\
    .groupby(['course_title', 'course_field', 'module_number', 'module_title', 'student_id'])\
    .agg({'start_date': np.min, 'finish_date': np.max})

# посчитаем всех студентов, которые начинали учитаться в каждом модуле
conversion_students_start = conversion_students.groupby(by=['course_title', 'course_field', 'module_number', 'module_title'])\
    ['start_date'].count().reset_index()

# посчитаем всех студентов, которые завершили обучение в каждом модуле
conversion_students_finish = conversion_students.dropna()\
    .groupby(by=['course_title', 'course_field', 'module_number', 'module_title'])\
    ['finish_date'].count().reset_index()

# объединим студентов, которые начинали учитаться в каждом модуле и которые завершили обучение в каждом модуле
conversion_students_all = conversion_students_start\
    .merge(conversion_students_finish, how='left',\
           left_on=['course_title', 'course_field', 'module_number', 'module_title'],\
           right_on=['course_title', 'course_field', 'module_number', 'module_title'])

#удаляем Nan из датасета
conversion_students_all = conversion_students_all.sort_values(['course_title', 'module_number']).dropna()

#сдвигаем значения столбца finish_date на 1 значение вниз
conversion_students_all['shift_finish_date'] = conversion_students_all['finish_date'].shift(1)

#считаем значение конверсии, на первом модуле курса конверсия равна 0
conversion_students_all['conversion'] = np.where(conversion_students_all['module_number'] == 1, 0, conversion_students_all['start_date']/conversion_students_all['shift_finish_date'])

# чистим ошибочное значение, задание в курсе начинаются не с первого модуля
conversion_students_all.loc[113, 'conversion'] = 0

# выведем первых 15 значений датасета
conversion_students_all.head(15)
```
